Mengubah kolom menjadi snake_case format

```python
covid_df.columns = covid_df.columns.str.lower().str.replace(' ', '_')
```

Mengecek kembali apakah semua nama kolom sudah sesuai dengan yang diinginkan yaitu snake_case format

```python
covid_df.info()
```

Mengganti kolom area_(km2) menjadi area_km2

```python
covid_df = covid_df.rename(columns={'area_(km2)': 'area_km2'})
```

Menghapus kolom 'city_or_regency' dikarenakan kolom ini tidak memiliki baris sama sekali

```python
covid_df = covid_df.drop(columns=['city_or_regency','special_status'])
```

Mengecek kembali tabel covid_df apakah kolom 'city_or_regency' sudah terhapus dan kolom 'area_(km2) sudah berganti nama jadi 'area_km2'

```python
covid_df.info()
```

Menghapus value 98 pada total_cities dikarenakan value tersebut anomali

```python
covid_df = covid_df[covid_df['total_cities'] != 98.0]
```

```python
covid_df['total_cities'].fillna(value=0, inplace=True) # mengganti column missing value pada total_cites menjadi 0 dikarenakan tidak ada kota madya pada provinsi tersebut
covid_df[covid_df.total_cities.isna()] # cek column missing value
```

Untuk mengisi kolom 'total_urban_villages' yang hilang, akan diterapkan metode mean untuk pengisian valuenya

```python
mean_value = covid_df['total_urban_villages'].mean()
covid_df['total_urban_villages'] = covid_df['total_urban_villages'].fillna(mean_value)
```

```python
covid_df.isna().sum()
```

Begitu juga untuk pengisian value kosong pada kolom 'total_rural_villages' akan dilakukan metode mean

```python
mean_value = covid_df['total_rural_villages'].mean()
covid_df['total_rural_villages'] = covid_df['total_rural_villages'].fillna(mean_value)
```

```python
covid_df.isna().sum()
```

```python
covid_df.growth_factor_of_new_cases.describe()
```

```python
new_case_diatas10 = covid_df[covid_df['growth_factor_of_new_cases'] > 10]['growth_factor_of_new_cases']
new_case_diatas10
```

Untuk pengisian value pada kolom 'growth_factor_of_new_cases' digunakan metode median

```python
median_value = covid_df['growth_factor_of_new_cases'].median()
covid_df['growth_factor_of_new_cases'] = covid_df['growth_factor_of_new_cases'].fillna(median_value)
```

```python
covid_df.isna().sum()
```

```python
covid_df.growth_factor_of_new_deaths.describe()
```

```python
new_death_diatas10 = covid_df[covid_df['growth_factor_of_new_deaths'] > 10]['growth_factor_of_new_deaths']
new_death_diatas10
```

Nilai '134.50' pada kolom 'growth_factor_of_new_deaths' sepertinya ini value anomali, jadi akan diubah menjadi 13.45

```python
covid_df.growth_factor_of_new_deaths.replace(covid_df.growth_factor_of_new_deaths.max(), 13.45, inplace=True)
```

```python
covid_df.growth_factor_of_new_deaths.describe()
```

Untuk pengisian value pada kolom 'growth_factor_of_new_deaths' digunakan metode median

```python
median_value = covid_df['growth_factor_of_new_deaths'].median()
covid_df['growth_factor_of_new_deaths'] = covid_df['growth_factor_of_new_deaths'].fillna(median_value)
```

```python
covid_df.isna().sum()
```

Ubah tipe data

```python
covid_df['date'] = pd.to_datetime(covid_df['date'])
covid_df['location_iso_code'] = covid_df['location_iso_code'].astype(str)
covid_df['location'] = covid_df['location'].astype(str)
covid_df['location_level'] = covid_df['location_level'].astype(str)
covid_df['province'] = covid_df['province'].astype(str)
covid_df['country'] = covid_df['country'].astype(str)
covid_df['continent'] = covid_df['continent'].astype(str)
covid_df['island'] = covid_df['island'].astype(str)


covid_df['case_fatality_rate'] = covid_df['case_fatality_rate'].str.replace('%', '').astype(float) / 100
covid_df['case_recovered_rate'] = covid_df['case_recovered_rate'].str.replace('%', '').astype(float) / 100
```

Export data

```python
covid_df.to_csv("covid_data.csv", index=False)
```
